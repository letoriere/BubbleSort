```python
def mergesort(list1):
    if len(list1)>1:
        mid = len(list1)//2
        lefthalf = list1[:mid]
        righthalf = list1[mid:]

        mergesort(lefthalf)
        mergesort(righthalf)
        merge(list1, lefthalf, righthalf)
```

```python
def merge(list1, lefthalf, righthalf):
        i=0
        j=0
        k=0
        
        while i < len(lefthalf) and j < len(righthalf):
            if lefthalf[i] < righthalf[j]:
                list1[k]=lefthalf[i]
                i=i+1
            else:
                list1[k]=righthalf[j]
                j=j+1
            k=k+1

        while i < len(lefthalf):
            list1[k]=lefthalf[i]
            i=i+1
            k=k+1

        while j < len(righthalf):
            list1[k]=righthalf[j]
            j=j+1
            k=k+1
```

```python
def verify(list1):
    for i in range(0,len(list1)-1,+1):
        if list1[i]>list1[i+1]:
            global verification
            verification = verification + 1
    if verification == 0:
        print('List is sorted.\n')
    else:
        print('List has not been sorted.\n')
```

```python
def test():
    start_time = time.time()
    mergesort(list1)
    print('The sorting took %s seconds.' % (time.time() - start_time))
```
