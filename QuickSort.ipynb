```python
def partition(list1,low,high):
    i = ( low-1 )
    pivot = list1[high]
 
    for j in range(low , high):
        if   list1[j] <= pivot:
            i = i+1
            list1[i],list1[j] = list1[j],list1[i]

    list1[i+1],list1[high] = list1[high],list1[i+1]
    return ( i+1 )
```

```python
def quicksort(list1,low,high):
    if low < high:

        pi = partition(list1,low,high)

        quicksort(list1, low, pi-1)
        quicksort(list1, pi+1, high)
```

```python
def verify(list1):
    for i in range(0,len(list1)-1,+1):
        if list1[i]>list1[i+1]:
            global verification
            verification = verification + 1
    if verification == 0:
        print('List is sorted.\n')
    else:
        print('List has not been sorted.\n')
```
