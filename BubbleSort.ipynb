```python
def bubblesort(list1):
    for index in range(len(list1)-1,0,-1):
         for i in range(index):
            if list1[i]>list1[i+1]:
                temp = list1[i]
                list1[i] = list1[i+1]
                list1[i+1] = temp
```

```python
def verify(list1):
    for index in range(len(list1)-1,0,-1):
        for i in range(index):
            if list1[i]>list1[i+1]:
                global verification
                verification = verification + 1
    if verification == 0:
        print('List is sorted.\n')
    else:
        print('List has not been sorted.\n')
```

```python
def test(bubblesort):
    start_time = time.time()
    bubblesort(list1)
    print('The sorting took %s seconds.' % (time.time() - start_time))
```

```python
import random
import time

n = (int)(input('N = '))
print('')

list1 = random.sample(range(-100000, 100000), n)
print('Unsorted list:')
print(list1)
print('')
```

```python
bubblesort(list1)
print('Sorted list:')
print(list1)
print('')
```

```python
verification = 0
verify(list1)
```

```python
test(bubblesort)
```
